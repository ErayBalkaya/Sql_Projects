# ÜRÜN-HER AYIN EN TREND ÜRÜNÜ

```python
sql=sql3 = """WITH monthly_sales AS (
    SELECT 
        EXTRACT(YEAR FROM o.order_date) AS year,
        EXTRACT(MONTH FROM o.order_date) AS month,
        p.product_name,
        c.category_name,  -- Directly from products table
        SUM(od.quantity) AS total_quantity,
        ROUND(SUM(od.unit_price * od.quantity * (1 - od.discount))::NUMERIC, 2) AS total_income
    FROM 
        orders o
    JOIN 
        order_details od ON o.order_id = od.order_id
    JOIN 
        products p ON od.product_id = p.product_id
	JOIN
		categories c ON c.category_id=p.category_id
    GROUP BY 
        EXTRACT(YEAR FROM o.order_date),
        EXTRACT(MONTH FROM o.order_date),
        p.product_name,
        c.category_name 
)
SELECT 
    year,
    month,
    product_name,
    category_name,  
    total_quantity,
    total_income
FROM (
    SELECT 
        year,
        month,
        product_name,
        category_name,  
        total_quantity,
        total_income,
        RANK() OVER (PARTITION BY year, month ORDER BY total_quantity DESC) AS rank
    FROM 
        monthly_sales
) ranked_sales
WHERE 
    rank = 1
ORDER BY 
    year, 
    month;"""
```

```python
df = sqlio.read_sql_query(sql,conn)
df.head()
```

```python
# Convert year and month to a single datetime column for plotting
df['date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))

# Create a plot
fig, ax = plt.subplots(figsize=(14, 8))

# Define width of bars and positions
bar_width = 0.4
bar_positions = range(len(df))

# Plot total income
bars1 = ax.barh(
    [p - bar_width / 2 for p in bar_positions], 
    df['total_income'], 
    bar_width, 
    color='b', 
    label='Total Income'
)

# Plot total quantity
bars2 = ax.barh(
    [p + bar_width / 2 for p in bar_positions], 
    df['total_quantity'], 
    bar_width, 
    color='r', 
    label='Total Quantity'
)

# Set labels and title
ax.set_yticks(bar_positions)
ax.set_yticklabels(df['date'].dt.strftime('%b %Y'))  # Set y-axis labels to month-year format
ax.set_xlabel('Value')
ax.set_title('Total Income and Total Quantity Sold per Month')

# Add product names on bars for total income
for barh, product_name in zip(bars1, df['product_name']):
    width = barh.get_width()
    ax.text(width + 50, barh.get_y() + barh.get_height() / 2.0, product_name,
            ha='left', va='center', fontsize=8, color='black')

# Add quantity values on bars for total quantity
for barh in bars2:
    width = barh.get_width()
    ax.text(width + 50, barh.get_y() + barh.get_height() / 2.0, f'{width:.0f}',
            ha='left', va='center', fontsize=8, color='black')

# Add legend and grid lines for better readability
ax.legend()
ax.grid(axis='x')

plt.tight_layout()
plt.show()
```

```python
# Aggregate total income and quantity by category_name
category_income = df.groupby('category_name')['total_income'].sum().reset_index()
category_quantity = df.groupby('category_name')['total_quantity'].sum().reset_index()

# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(18, 8))

# Pie chart for total income
ax[0].pie(category_income['total_income'], labels=category_income['category_name'], autopct='%1.1f%%', startangle=140)
ax[0].set_title('Income Distribution by Category', fontsize=16, fontweight='bold')

# Pie chart for total quantity
ax[1].pie(category_quantity['total_quantity'], labels=category_quantity['category_name'], autopct='%1.1f%%', startangle=130)
ax[1].set_title('Quantity Distribution by Category', fontsize=16, fontweight='bold')

# Equal aspect ratio ensures that pie is drawn as a circle.
ax[0].axis('equal')
ax[1].axis('equal')

plt.show()
```

# FİYAT - ÜRÜNLERİN KATEGORİ BAZLI FİYAT DEĞİŞİM ORANLARI

```python
sql2="""WITH monthly_prices AS (
    SELECT
        EXTRACT(YEAR FROM o.order_date) AS year,
        EXTRACT(MONTH FROM o.order_date) AS month,
        c.category_name,
        AVG(od.unit_price) AS avg_price
    FROM
        orders o
    JOIN
        order_details od ON o.order_id = od.order_id
    JOIN
        products p ON od.product_id = p.product_id
    JOIN
        categories c ON p.category_id = c.category_id
    GROUP BY
        EXTRACT(YEAR FROM o.order_date),
        EXTRACT(MONTH FROM o.order_date),
        c.category_name
),
price_changes AS (
    SELECT
        mp1.year,
        mp1.month,
        mp1.category_name,
        mp1.avg_price AS current_price,
        LAG(mp1.avg_price) OVER (PARTITION BY mp1.category_name ORDER BY mp1.year, mp1.month) AS previous_price
    FROM
        monthly_prices mp1
)
SELECT
    pc.year,
    pc.month,
    pc.category_name,
    pc.current_price,
    pc.previous_price,
    CASE
        WHEN pc.previous_price IS NULL THEN NULL
        ELSE ((pc.current_price - pc.previous_price) / pc.previous_price) * 100
    END AS price_change_percentage
FROM
    price_changes pc
WHERE
    pc.previous_price IS NOT NULL
    AND ((pc.current_price - pc.previous_price) / pc.previous_price) * 100 <> 0
ORDER BY
    pc.category_name,
    pc.year,
    pc.month;
    """
```

```python
df = sqlio.read_sql_query(sql2,conn)
df.head()
```

```python
# Create a new column for formatted date to help with plotting
df['formatted_date'] = pd.to_datetime(df[['year', 'month']].assign(day=1)).dt.to_period('M').astype(str)  # Format to 'YYYY-MM'

# Pivot data for heatmap
heatmap_data = df.pivot_table(index='formatted_date', columns='category_name', values='price_change_percentage', aggfunc='mean')

# Plotting the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f")
plt.xlabel('Category Name')
plt.ylabel('Year-Month')
plt.title('Price Change Percentage Heatmap')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create subplots for each category
categories = df['category_name'].unique()
num_categories = len(categories)

# Create a grid of subplots
fig, axes = plt.subplots(nrows=(num_categories // 2) + (num_categories % 2), ncols=2, figsize=(16, 5 * ((num_categories // 2) + (num_categories % 2))))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plotting each category's price change in separate charts
for i, category in enumerate(categories):
    category_data = df[df['category_name'] == category]
    ax = axes[i]
    ax.plot(category_data['year'].astype(str) + '-' + category_data['month'].astype(str).str.zfill(2),
            category_data['price_change_percentage'],
            marker='o', label=category)
    ax.set_title(f'Price Change for {category}')
    ax.set_xlabel('Year-Month')
    ax.set_ylabel('Price Change Percentage')
    ax.legend()
    ax.tick_params(axis='x', rotation=45)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
